# HTML

## Qué es HTML?
- **(HyperText Markup Language)**
- es el lenguaje estándar para definir la **estructura** de páginas web. 
- Una página HTML es un **árbol** de elementos: una **raíz** `<html>`
  - nodos hijos:
    - `<head>`
    - `<body>`
  
![](html.png)

## Estructura base de un documento
- inciar con `<!DOCTYPE html>` (indoca que es HTML5) y contiene la raiz `<html>` con dos secciones:
  - `<head>`: metadatos (título, estilos, scripts, idioma, etc)
  - `<body>`: contenido visible

````python
<!DOCTYPE html>
<html lang="es">
    <head>
        <title> Mi Primera Página Web </title>
    </head>
    <body>
        <h1> Hola Mundo </h1>
        <p> Esta es mi primera página web. </p>
    </body>
</html>"
````

- `<!DOCTYPE html>`: indica el estandar del documento
- `<html lang="es">`: raiz del docuemtno y el idioma
- `<head>`: metadatos
- `<body>`: contenido visible

## Texto
- **Título**:  `<h1>` a `<h6>`. De mayor a mejor importancia
- **Párrafos**: `<p>` para bloques de texto.
- **Salto de línea**: `<br>`
- **Separador**: `<hr>` para una línea horizontal.


````python
<!DOCTYPE html>
<html lang="es">
    <head>
        <meta charset="UTF-8">
        <title> Mi Primera Página Web </title>
    </head>
    <body>
        <header>

            <h1> Mi Primera Página Web </h1>

            <nav>
                <a href="bienvenida">Bienvenida</a>
                <a href="sobre-mi">Sobre mí</a>
                <a href="contacto">Contacto</a>
            </nav>

        </header>

        <main id="inicio">
            <section id="bienvenida">
                <h2> Bienvenido a mi sitio web </h2>
                <p> Este es un sitio de prueba para aprender HTML. </p>
            </section>

            <section id="sobre-mi">
                <h2> Sobre mí </h2>
                <p> Me llamo Diego y estoy aprendiendo a programar. </p>
            </section>

        </main>

        <footer id="contacto">
            <p> Contacto: <a href="mailto:ejemplo@email.com">ejemplo@email.com</a> </p>
            <p> Derechos reservados 2026 </p>

        </footer>
    </body>
</html>

````

## formularios
- `<form action="... " method="get|post">` agrupa campos.  
- `<label for="id">` vincula texto a un control con `id`.  
- `<input>` soporta múltiples **type**: `text`, `email`, `password`, `radio`, `checkbox`, `file`, `submit`, etc.  
-  El atributo **`name`** es la **clave** que viaja al servido.
-   `placeholder` sugiere el contenido.  
-   `<button type="button|reset|submit">` botones personalizables.
-   `<textarea rows="..." cols="...">` para texto multilínea.

## Listas (`<ul>`, `<ol>`, `<li>`)
- **No ordenada (`<ul>`)**: viñetas.  
- **Ordenada (`<ol>`)**: numeración.  
- Se pueden **anidar** listas;
- se puede cambiar el marcador con CSS (`list-style-type`).


````python

<h3>Lista no ordenada</h3>
<ul>
  <li>Elemento 1</li>
  <li>Elemento 2</li>
</ul>

<h3>Lista ordenada</h3>
<ol>
  <li>Paso A</li>
  <li>Paso B</li>
</ol>

<h3>Lista anidada</h3>
<ol>
  <li>Tema 1
    <ol style="list-style-type: lower-roman;">
      <li>subtema i</li>
      <li>subtema ii</li>
    </ol>
  </li>
  <li>Tema 2</li>
</ol>
````

## CSS
Toda regla se compone estrictamente de tres partes:

1. **Selector:** El elemento HTML que queremos afectar.
2. **Propiedad:** El atributo visual que vamos a cambiar (color, tamaño, bordes).
3. **Valor:** El nuevo aspecto que le daremos a esa propiedad.

```css
selector {
    propiedad: valor;
}

```

### Las 3 Formas de Aplicar CSS en un Proyecto Web
#### Forma 1: CSS en Línea (Atributo `style`)
```html
<h1 style="color: #204ecf; font-size: 24px;">Estación Órbita Alfa</h1>
<p style="color: #333333;">Iniciando protocolo de hangar...</p>

```

#### Forma 2: CSS Interno o Embebido (Etiqueta `<style>`)
```html
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>Panel de Control</title>
    <style>
        body { 
            font-family: Arial, sans-serif; 
            background: #f6f6f6; 
            margin: 24px; 
        }
        h1 { 
            color: #204ecf; 
        }
        a { 
            text-decoration: none; 
        }
        .card { 
            background: white; 
            padding: 16px; 
            border-radius: 8px; 
            box-shadow: 0 2px 8px rgba(0,0,0,.08); 
        }
    </style>
</head>
<body>
    <div class="card">
        <h1>Misión Autorizada</h1>
        <a href="/">Volver al panel</a>
    </div>
</body>
</html>

```

#### Forma 3: CSS Externo (Archivos `.css` independientes)
##### Paso 1: Crear el archivo de estilos (ej. `public/css/estilos.css`)
```css
        body { 
            font-family: Arial, sans-serif; 
            background: #f6f6f6; 
            margin: 24px; 
        }
        h1 { 
            color: #204ecf; 
        }
        a { 
            text-decoration: none; 
        }
        .card { 
            background: white; 
            padding: 16px; 
            border-radius: 8px; 
            box-shadow: 0 2px 8px rgba(0,0,0,.08); 
        }
```

##### Paso 2: Vincular el archivo en el HTML (`public/index.html`)

Usamos la etiqueta `<link>` dentro del `<head>`. El atributo `href` debe apuntar a la ruta del archivo CSS.
```html
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>Panel de Control - Externo</title>
    <link rel="stylesheet" href="css/estilos.css">
</head>
<body>
    <div class="card">
        <h1>Conexión Exitosa</h1>
    </div>
</body>
</html>

```

## Los selectores CSS
### Selectores de etiqueta
Afecta a **todas** las etiquetas de ese tipo en el documento.

* **Sintaxis:** Se escribe el nombre de la etiqueta directamente.
* **Código:**
```css
h1 {
    color: #204ecf;
}


### Selector de Clase (El punto: `.nombre`)

```css
.card {
    background: white;
    border-radius: 8px;
}

```

### Selector de ID  (El numeral: `#nombre`)
```css
#boton-despacho {
    background: green;
}

```
* **Código en HTML:**
```html
<button id="boton-despacho">Despachar</button>

```


## Diccionario de Propiedades Básicas de Diseño

### Propiedades de Color y Fondo

* **`color`:** Cambia el color del texto. Puede usar nombres (`red`), códigos hexadecimales (`#204ecf`) o RGB (`rgb(32, 78, 207)`).
* **`background-color` (o `background`):** Cambia el color de fondo del contenedor.

### Propiedades de Tamaño y Tipografía

* **`font-size`:** Define el tamaño de la letra. Se mide comúnmente en píxeles (`px`). Ejemplo: `font-size: 16px;`.
* **`font-family`:** Define la tipografía. Ejemplo: `font-family: Arial, sans-serif;`.
* **`width` y `height`:** Controlan el ancho y el alto de un contenedor o imagen. Ejemplo: `width: 300px; height: 150px;`.

### Propiedades de Forma y Bordes

* **`border`:** Crea una línea alrededor del elemento. Requiere tres valores: grosor, estilo y color. Ejemplo: `border: 2px solid #000000;` (Línea sólida negra de 2 píxeles).
* **`border-radius`:** Redondea las esquinas de los contenedores para darles un aspecto moderno de "tarjeta". Ejemplo: `border-radius: 12px;`.

### Propiedades de Espaciado (El Modelo de Caja)
Todo elemento en HTML es una caja. El espacio se controla con dos propiedades vitales:

* **`padding` (Espaciado Interno):** Es la distancia entre el texto y el borde de su propio contenedor.
* **`margin` (Espaciado Externo):** Es la distancia de separación entre este contenedor y los otros elementos que tiene afuera.
